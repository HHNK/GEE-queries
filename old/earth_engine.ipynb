```python
import math
import sys
# var palettes = require('users/gena/packages:palettes');
# var palette = palettes.colorbrewer.PRGn [11];
daterange = ("2018-06-01", "2018-07-01")
daterangestr = "{}_{}".format(daterange[0], daterange[1])

# sentinel3 data
dataset = ee.ImageCollection('COPERNICUS/S3/OLCI').filterBounds(ijssel_rect).filterDate(daterange[0], daterange[1])

# dates if needed
# var now = Date.now();
# var end = ee.Date(now).format();

def getQABits(image, start, end, newName):
    # Compute the bits we need to extract.
    pattern = 0;
    for i in range(start,end):
        pattern += math.pow(2, i)
    
    # Return a single band image of the extracted QA bits, giving the band
    # a new name.
    return image.select([0], [newName]).bitwiseAnd(pattern).rightShift(start)


def get_MCI_clipped(img):
    """convert ee.Image to MCI ee.Image"""
    # var img = dataset.filterDate(range.start(), range.end()).first()
    # img = img.clip(ijssel_precise)

    img = img.clip(ijssel_rect);
    # var date = ee.Date(img.get('system:time_start'));
    # var date_string = date.format('dd-MM-YYYY_HH-mm-ss');

    # get bright pixels
    QA = img.select(['quality_flags']);
    bright_pix = getQABits(QA, 24,27, 'Bright').eq(0);
  
    # calculate MCI
    # bands can be found here: https://developers.google.com/earth-engine/datasets/catalog/COPERNICUS_S3_OLCI
    # l709 = img.select('Oa11_radiance');
    # l681 = img.select('Oa10_radiance');
    # l753 = img.select('Oa12_radiance');

    MCI = img.expression(
        'l709 - l681 - (((709-681)/(753-681)) * (l753-l681))', {
        'l709': img.select('Oa11_radiance'),
        'l681': img.select('Oa10_radiance'),
        'l753': img.select('Oa12_radiance')
    })
    
#     MCI = MCI.toByte()
#     bright_pix = ee.Image(bright_pix)
  
#     MCI_cloud_clip = MCI.updateMask(bright_pix);
  
    MCI = MCI.copyProperties(img)
  
    return MCI


def get_RGB(img):
    # var img = dataset.filterDate(range.start(), range.end()).first()
    # img = img.clip(ijssel_precise)

    img = img.clip(ijssel_rect)

    # get bright pixels
#     QA = img.select(['quality_flags']);
#     bright_pix = getQABits(QA, 24,27, 'Bright').eq(0)
#     print(bright_pix.getInfo())


    # Convert to radiance units, conversion constants can be found here as well: https://developers.google.com/earth-engine/datasets/catalog/COPERNICUS_S3_OLCI
    rgb = img.select(['Oa08_radiance', 'Oa06_radiance', 'Oa04_radiance']).multiply(ee.Image([0.00876539, 0.0123538, 0.0115198]))

#     mask = img.expression(
#         'l709 - QA * 10', {
#         'l709': img.select('Oa08_radiance'),
#         'QA': QA
#     })
    
#     rgb = rgb.updateMask(mask);
#     rgb = rgb.copyProperties(img)
    return rgb


RGB = dataset.map(get_RGB)
MCI_clipped = dataset.map(get_MCI_clipped)

# cloud_mask_clipped = dataset.map(get_cloudmask_clipped)

dates = dataset.aggregate_array("system:time_start")

def extract_date(image):
    return ee.Feature(None, {"date": image.date().format('YYYY-MM-dd')})

dates = dataset.map(extract_date).aggregate_array('date')
  
datelist = ee.List(dates)
len_images = datelist.length().getInfo()
print("total images found")
print(len_images)


listy = MCI_clipped.toList(len_images)
for i in range(len_images):
    try:
        image=ee.Image(listy.get(i));
        date = datelist.get(i).getInfo()
        name = 'MCI_' + date;
        print(name);
        # Export MCI
        batch.Download.image.toLocal(image, name=name, scale=300, region=ijssel_rect, path='exports/{}/MCI/'.format(daterangestr), toFolder=True)
    except ValueError:
        print('something went wrong in ' + str(i))


listy = RGB.toList(len_images)
for i in range(len_images):
    try:
        image=ee.Image(listy.get(i));
        date = datelist.get(i).getInfo()
        name = 'RGB_' + date;
        print(name);
        # Export MCI
        batch.Download.image.toLocal(image, name=name, scale=300, region=ijssel_rect, path='exports/{}/RGB/'.format(daterangestr), toFolder=True)
    except ValueError:
        print('something went wrong in ' + str(i))
```
